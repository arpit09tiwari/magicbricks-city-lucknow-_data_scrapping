```python
df = df_orig.copy()
df['carpet_area'] = df['carpet_area'].apply(lambda x: x.replace('sqft', '')
                                if isinstance(x, str) else x).astype(float)
df['price'] = df['price'].apply(lambda x: x.replace('₹', '').replace('Lac', '').replace('Cr', '')
                                if isinstance(x, str) else x).astype(float)
df.loc[df['price'] < 10, 'price'] = df['price']*100
df
```
